```python
import os
def compare_files(file1_path, file2_path):
    if not os.path.isfile(file1_path):
        print(f"Error: File '{file1_path}' tidak ditemukan.")
        return
    if not os.path.isfile(file2_path):
        print(f"Error: File '{file2_path}' tidak ditemukan.")
        return

    with open(file1_path, 'r', encoding='utf-8') as file1, open(file2_path, 'r', encoding='utf-8') as file2:
        lines1 = file1.readlines()
        lines2 = file2.readlines()

    max_lines = max(len(lines1), len(lines2))

    print("Perbandingan isi file:\n")

    berbeda = False
    for i in range(max_lines):
        line1 = lines1[i].rstrip('\n') if i < len(lines1) else '[Tidak ada baris]'
        line2 = lines2[i].rstrip('\n') if i < len(lines2) else '[Tidak ada baris]'

        if line1 != line2:
            berbeda = True
            print(f"Baris {i+1}:")
            print(f"  File 1: {line1}")
            print(f"  File 2: {line2}")
            print()

    if not berbeda:
        print("Kedua file identik (tidak ada perbedaan).")

file1 = "daftarbuku.txt"
file2 = "mbox-short.txt"

compare_files(file1, file2)
```

```python
def kuis_dari_file(nama_file):
    print(f"nama file1: {nama_file}")
    
    try:
        with open(nama_file, 'r', encoding='utf-8') as file:
            for line in file:
                if '||' in line:
                    soal, jawaban_benar = line.strip().split('||')
                    soal = soal.strip()
                    jawaban_benar = jawaban_benar.strip()

                    print(soal)
                    jawaban_user = input("Jawab: ").strip()

                    if jawaban_user.lower() == jawaban_benar.lower():
                        print("Jawaban benar!\n")
                    else:
                        print("Jawaban salah!\n")
                else:
                    print("Format soal tidak valid.\n")

    except FileNotFoundError:
        print(f"File '{nama_file}' tidak ditemukan.")

# Jalankan program
kuis_dari_file("soal.txt")
```
